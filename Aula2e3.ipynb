```python
import pandas as pd
import plotly.express as px
```

```python
df_principal = pd.read_excel("/content/Python.xlsx", sheet_name="Principal")
df_principal
```

# Nova seção

```python
df_ticker = pd.read_excel("/content/Python.xlsx", sheet_name="Ticker")
df_ticker
```

```python
df_totaldeacoes = pd.read_excel("/content/Python.xlsx", sheet_name="Total_de_acoes")
df_totaldeacoes
```

```python
df_chatgpt = pd.read_excel("/content/Python.xlsx", sheet_name="Chatgpt")
df_chatgpt
```

```python
df_principal = df_principal[['Ativo', 'Data', 'Último (R$)', 'Var. Dia (%)']].copy()
df_principal
```

```python
df_principal = df_principal.rename(columns={'Último (R$)': 'Valor_Final', 'Var. Dia (%)':'Var_dia_pct'}).copy()
df_principal
```

```python
df_principal['Var_pct'] = df_principal['Var_dia_pct'] / 100
df_principal['Valor_inicial'] = df_principal['Valor_Final'] / (df_principal['Var_pct'] + 1)
df_principal
```

```python
df_principal = df_principal.merge(df_totaldeacoes, left_on='Ativo', right_on='Código', how='left')
df_principal
```

```python
df_principal = df_principal.drop(columns=['Código'])
df_principal
```

```python
df_principal['Variacao_rs']= (df_principal['Valor_Final'] - df_principal['Valor_inicial'])*df_principal['Qtde. Teórica']
df_principal
```

```python
pd.options.display.float_format = '{:.2f}'.format
```

```python
df_principal['Qtde. Teórica'] = df_principal['Qtde. Teórica'].astype(int)
df_principal
```

```python
df_principal = df_principal.rename(columns={'Qtde_teorica':'Qtd_teorica'}).copy()
df_principal
```

```python
df_principal['Resultado'] = df_principal['Variacao_rs'].apply(lambda x: "Subiu" if x > 0 else ("Desceu" if x < 0 else "Estável"))
df_principal
```

```python
df_principal = df_principal.merge(df_ticker, left_on='Ativo', right_on='Ticker', how='left')
df_principal = df_principal.drop(columns=['Ticker'])
df_principal
```

```python
df_principal = df_principal.merge(df_chatgpt, left_on='Nome_x', right_on='Nome da Empresa', how='left')
df_principal = df_principal.drop(columns=['Nome da Empresa'])
df_principal
```

```python
df_principal['Cat_idade'] = df_principal['Idade em Anos'].apply(lambda x: "Mais de 100 anos" if x > 100 else ("Menos de 50 anos" if x < 50 else "Entre 50 e 100 anos"))
df_principal
```

```python
# Calcular o valor máximo, mínimo e média da coluna 'Variacao_rs'
maior_valor = df_principal['Variacao_rs'].max()
menor_valor = df_principal['Variacao_rs'].min()
media_total = df_principal['Variacao_rs'].mean()

# Calcular a média para cada grupo separado
media_subiu = df_principal[df_principal['Resultado'] == 'Subiu']['Variacao_rs'].mean()
media_desceu = df_principal[df_principal['Resultado'] == 'Desceu']['Variacao_rs'].mean()

# Mostrar os resultados
print("Maior\t", f"R$ {maior_valor:,.2f}")
print("Menor\t", f"R$ {menor_valor:,.2f}")
print("Média\t", f"R$ {media_total:,.2f}")
print("Média de quem subiu\t", f"R$ {media_subiu:,.2f}")
print("Média de quem desceu\t", f"R$ {media_desceu:,.2f}")
```

```python
df_principal_subiu = df_principal[df_principal['Resultado'] == 'Desceu']
df_principal_subiu
```

```python
df_analise_segmento = df_principal_subiu.groupby('Segmento')['Variacao_rs'].sum().reset_index()
df_analise_segmento
```

```python
df_analise_saldo = df_principal.groupby('Resultado')['Variacao_rs'].sum().reset_index()
df_analise_saldo
```

```python
fig = px.bar(df_analise_saldo, x='Resultado', y='Variacao_rs', text='Variacao_rs', title='Variação Reais por Resultado')
fig.show()
```

```python
fig.update_traces(texttemplate='R$ %{text:,.2f}', textposition='outside')
fig.show()
```

```python
fig = px.pie(df_analise_segmento, names='Segmento', values='Variacao_rs', title='Variação Reais por Segmento')
fig.show()
```

```python
df_cat_idade = df_principal.groupby('Cat_idade')['Variacao_rs'].sum().reset_index()
df_cat_idade
```

```python
fig = px.bar(df_cat_idade, x='Cat_idade', y='Variacao_rs', text='Variacao_rs', title='Variação Reais por Categoria de Idade')
fig.show()
```
